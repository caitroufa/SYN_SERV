# Introduction to Working With Dataframes
Working with tabular data is highly important when it comes to data science. There are a few a different methods of working with this data, but the most commonly used method uses the Python library ```pandas```. This notebook will hopefully help to become more familiar with ```pandas``` and all it has to offer. 

````python
#Import the libraries we want to use
import pandas as pd
import numpy as np
````

### Loading a File and Reviewing the Basics

````python
#load a csv as a pandas dataframe.
df = pd.read_csv('storm_modes_1992_hist.csv')
df
````

````python
#See all the unique entries in a column.
#in this case, list all the labels of convective mode.
df['Mode'].unique()
````

````python
#seperate timing values into their own column
#this is a good method for adding columns to a dataframe
df['Year'] = pd.to_datetime(df['Time']).dt.year
df['Month'] = pd.to_datetime(df['Time']).dt.month
df['Day'] = pd.to_datetime(df['Time']).dt.day
df['Hour'] = pd.to_datetime(df['Time']).dt.time
````

````python
# delete the old time and index column
df = df.drop(columns=['Time','Unnamed: 0'])
df
````

### Groupby
Grouping the data can help create links that need to be made.

````python
#group the data by Mode and find the mean value of the dbz and uh
df.groupby(['Mode'])['refc','uh'].mean()
````

````python
#group the data by Mode and Month and find the max value of dbz and uh
df.groupby(['Mode', 'Month'])['refc','uh'].max()
````

````python
#find the number of events in each month
df.groupby(['Month'])['Mode'].count()
````

### GeoViz with Pandas (and Matplotlib)

````python
#Generate a quick histogram from the data
df.Mode.hist()
````

````python
#show the frequency of each mode with respect to the month
df.groupby(['Mode','Month']).size().unstack().plot(kind='bar', stacked=True)
````

````python
#of course we can make things a little fancier if needed
#matplotlib is helpful for making better plots
import matplotlib.pyplot as plt
from matplotlib import rc

#save off the data we want from the dataframe
data = df.groupby(['Month','Mode']).size().unstack()
data = data.fillna(0)
data
````

````python
# create a zeros array to create a bottom value
bottom = np.zeros(data[data.index == 1].values[0].shape)

# create a list of hex codes to use to color the array
colors = ['#f9e442','#f7ce63','#f3aa79','#f0959e','#ee8cb5','#c693be','#937dc0','#5fa3d9','#00b2e2','#54bcbb','#69bca8','#8dc05a']

# forloop to iterate through the months
for i in range(len(data.index)):
    
    # plot the values
    plt.bar(data.columns.values, data[data.index == i+1].values[0], color=colors[i], edgecolor='white', bottom=bottom, width=0.8)
    
    # add the current month's values to the bottom value
    bottom = bottom + data[data.index == i+1].values[0]
    
    # create legend
    plt.legend(colors, labels=data.index.values)
````
