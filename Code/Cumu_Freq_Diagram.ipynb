```python
#import packages
import pandas as pd
import glob
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import matplotlib.dates as mdates
```

```python
#threshold all severe in HIST, then assign datetime
df_HIST = df_HIST[((df_HIST['DBZ'] >= 40.0) & (df_HIST['UVV'] >= 25.0))]
df_HIST['Time'] = pd.to_datetime(df_HIST['Time'])
df_HIST['Date'] = pd.to_datetime(df_HIST['Time']).dt.date
df_HIST['Year'] = pd.to_datetime(df_HIST['Time']).dt.year
df_HIST['Month'] = pd.to_datetime(df_HIST['Time']).dt.month
df_HIST
```

```python
#as in previous cell, but for 8.5 end
df_8p5END = df_8p5END[((df_8p5END['DBZ'] >= 40.0) & (df_8p5END['UVV'] >= 25.0))]
df_8p5END['Time'] = pd.to_datetime(df_8p5END['Time'])
df_8p5END['Date'] = pd.to_datetime(df_8p5END['Time']).dt.date
df_8p5END['Year'] = pd.to_datetime(df_8p5END['Time']).dt.year
df_8p5END['Month'] = pd.to_datetime(df_8p5END['Time']).dt.month
df_8p5END
```

```python
#function to convert water year to calendar year and output daily count
def Cumu_sum(df):

    daily_counts = pd.DataFrame(index=np.arange(1,366,1))
    df['Day_of_Year'] = pd.to_datetime(df['Date']).dt.dayofyear
    
    for year in df['Year'].unique()[:-1]:
        b=[]
        
        if year == np.min(df['Year'].unique()):
            df_year = df[((df['Year'] == np.min(df['Year'].unique())) | (df['Year'] == np.max(df['Year'].unique())))]
            
            for day in np.arange(1,366,1):
                df_sub = df_year[(df_year['Day_of_Year'] == day)]
                b.append(len(df_sub))
            
            
        else:
            df_year = df[(df['Year'] == year)]
            
            for day in np.arange(1,366,1):
                df_sub = df_year[(df_year['Day_of_Year'] == day)]
                b.append(len(df_sub))
            
        daily_counts[year] = b
            
    return daily_counts
```

```python
#calculate daily counts for each epoch and convert to calendar year
new_df_HIST = Cumu_sum(df_HIST)
new_df_4p5MID = Cumu_sum(df_4p5MID)
new_df_8p5MID = Cumu_sum(df_8p5MID)
new_df_4p5END = Cumu_sum(df_4p5END)
new_df_8p5END = Cumu_sum(df_8p5END)
```

```python
#figure specifications and plotting
fig, ax = plt.subplots(figsize = (6, 6), dpi = 300)

ax.set_title('Cumulative HCW Counts')
ax.set_xlabel('Month')
ax.set_ylabel('Counts')
ax.grid(lw = 0.5, alpha = 0.5, zorder = 0.1)

#hist
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_HIST.mean(axis = 1)), c = 'black', lw = 2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_HIST.mean(axis = 1)), c = '#c5eabc', label = 'HIST')

#mid
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5MID.mean(axis = 1)), c = 'black', lw = 2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5MID.mean(axis = 1)), c = '#b5d6ea', label = 'MID4.5')
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5MID.mean(axis = 1)), c = 'black', lw = 2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5MID.mean(axis = 1)), c = '#ca6e74', label = 'MID8.5')

#end
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5END.mean(axis = 1)), c = 'black', lw = 2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5END.mean(axis = 1)), c = '#006db3', label = 'END4.5')
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5END.mean(axis = 1)), c = 'black', lw = 2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5END.mean(axis = 1)), c = '#a51017', label = 'END8.5')

#setting x-axis ticks to months
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

ax.set_xlim(0, 364)
ax.legend()
plt.tight_layout()
#plt.savefig('Cumulative Frequency.png', dpi = 300, bbox_inches='tight')
```

```python
# function to calc severe days and convert from water year to calendar year
def Cumu_days(df):
    daily_counts = pd.DataFrame(index=np.arange(1, 366, 1))
    df['Day_of_Year'] = pd.to_datetime(df['Date']).dt.dayofyear
    
    for year in df['Year'].unique()[:-1]:
        b = []
        
        if year == np.min(df['Year'].unique()):
            df_year = df[((df['Year'] == np.min(df['Year'].unique())) | (df['Year'] == np.max(df['Year'].unique())))]
            
            for day in np.arange(1, 366, 1):
                # if severe on a day, count as 1, otherwise 0
                df_sub = df_year[(df_year['Day_of_Year'] == day)]
                b.append(1 if len(df_sub) > 0 else 0)
                
        else:
            df_year = df[(df['Year'] == year)]
            
            for day in np.arange(1, 366, 1):
                df_sub = df_year[(df_year['Day_of_Year'] == day)]
                b.append(1 if len(df_sub) > 0 else 0)
            
        daily_counts[year] = b
            
    return daily_counts
```

```python
# calculate for each epoch using new days function
new_df_HIST_days = Cumu_days(df_HIST)
new_df_4p5MID_days = Cumu_days(df_4p5MID)
new_df_8p5MID_days = Cumu_days(df_8p5MID)
new_df_4p5END_days = Cumu_days(df_4p5END)
new_df_8p5END_days = Cumu_days(df_8p5END)
```

```python
# plot cumulative frequency of severe weather days
fig, ax = plt.subplots(figsize=(6, 6), dpi=300)

ax.set_title('Cumulative HCW Days')
ax.set_xlabel('Month')
ax.set_ylabel('Days')
ax.grid(lw=0.5, alpha=0.5, zorder=0.1)

# plot for HIST
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_HIST_days.mean(axis=1)), c='black', lw=2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_HIST_days.mean(axis=1)), c='#c5eabc', label='HIST')

# plot for MID 4.5 and 8.5
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5MID_days.mean(axis=1)), c='black', lw=2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5MID_days.mean(axis=1)), c='#b5d6ea', label='MID4.5')
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5MID_days.mean(axis=1)), c='black', lw=2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5MID_days.mean(axis=1)), c='#ca6e74', label='MID8.5')

# plot for END 4.5 and 8.5
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5END_days.mean(axis=1)), c='black', lw=2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_4p5END_days.mean(axis=1)), c='#006db3', label='END4.5')
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5END_days.mean(axis=1)), c='black', lw=2.5)
ax.plot(np.arange(1, 366, 1), np.cumsum(new_df_8p5END_days.mean(axis=1)), c='#a51017', label='END8.5')

# set x-axis ticks to months
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

ax.set_xlim(0, 364)
ax.legend()
plt.tight_layout()
# plt.savefig('Cumulative_HCW_Days.png', dpi=300, bbox_inches='tight')

plt.show()
```
